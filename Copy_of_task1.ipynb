```python
import time
import dask.dataframe as dd
import pandas as pd

start_time = time.time()

file_path = '/content/drive/MyDrive/Classroom/12 th dilligent 1st/dataset(Task1). csv'
# Load the csv using dask.
# Specify dtypes to avoid type inference issues, use float64 for numerical columns and object for strings
df_dask = dd.read_csv(
    file_path,
    encoding='ISO-8859-1',
    dtype={'InvoiceNo': 'object',
           'StockCode': 'object',
           'Description': 'object',
           'Quantity': 'int64',
           'InvoiceDate': 'object',
           'UnitPrice': 'float64',
           'CustomerID':'float64',
           'Country':'object'})

# Example operation (make it a more complex operation so we can see potential dask benefits)
# Example: apply a function to a column
df_dask["new_column"] = df_dask['UnitPrice'] * df_dask['Quantity']

# Use a smaller computation to show that the .compute() call is what takes so long
df_dask_sample = df_dask.head()
# Compute a sample, this is much faster than computing everything

end_time = time.time()
print(f"Dask Execution Time: {end_time - start_time:.2f} seconds")
```

```python
from dask.diagnostics import ProgressBar
import dask.dataframe as dd
import pandas as pd

# Assuming df_dask_large is already defined from previous cells.
# If not, recreate it like this:
df_large = pd.concat([df_pandas] * 10, ignore_index=True)
output_file_path = "/content/large_dataset.csv"
df_large.to_csv(output_file_path, index=False)
df_dask_large = dd.read_csv(output_file_path, dtype={
    'InvoiceNo': 'object',
    'StockCode': 'object',
    'Description': 'object',
    'Quantity': 'int64',
    'InvoiceDate': 'object',
    'UnitPrice': 'float64',
    'CustomerID': 'float64',
    'Country': 'object',
    'new_column':'object'
})

with ProgressBar():
    result = df_dask_large.groupby(df_dask_large.columns[0]).mean(numeric_only=True).compute()
print(result)
```
